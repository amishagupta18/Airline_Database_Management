### Building the 'Airplane Model' table: 

```python
airplane_model = fleet_df[['Aircraft Type', 'Unit Cost']]
airplane_model = airplane_model.dropna()
airplane_model = airplane_model.drop_duplicates()
airplane_model['Unit Cost'] = airplane_model['Unit Cost'].str.replace('$', '')
airplane_model['Unit Cost'] = airplane_model['Unit Cost'].astype(float)
airplane_model.info()
```

```python
airplane_model = airplane_model.groupby('Aircraft Type', as_index=False).mean()
airplane_model.to_csv('airplane_model', index=False)
```

### Building the 'Airline Company' table:

```python
airline_company = fleet_df[['Airline', 'Parent Airline']]
airline_company = airline_company.drop_duplicates()
data = tail_df[['Airline']]
```

### Building the 'Passengers' table:

```python
customer_id = []
for i in range(1, 301):
    customer_id.append(''.join(random.choice('0123456789ABCDEF') for i in range(6)))
```

```python
new_names = []
for i in range(1, 301):
    new_names.append(names.get_full_name())
```

```python
def gen_phone():
    first = str(random.randint(100,999))
    second = str(random.randint(1,888)).zfill(3)

    last = (str(random.randint(1,9998)).zfill(4))
    while last in ['1111','2222','3333','4444','5555','6666','7777','8888']:
        last = (str(random.randint(1,9998)).zfill(4))
        
    return '{}-{}-{}'.format(first,second, last)

phone_numbers = []
for _ in range(300):
    phone_numbers.append(gen_phone())
```

```python
dict = {'Customer ID': customer_id, 'Customer Name': new_names, 'Phone Number': phone_numbers}
passengers = pd.DataFrame(dict)
passengers.to_csv('passengers', index=False)
```
